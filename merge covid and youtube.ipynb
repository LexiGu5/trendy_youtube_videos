```python
import pandas as pd
```

```python
# load the data files 
df = pd.read_csv('trending_youtube.csv')
covid = pd.read_csv('covid.csv')
```

```python
# rename the column for merging 
covid.rename(columns = {'Date':'publishedAt'}, inplace = True)
```

```python
# convert object to datetime to make the data type consistent 
covid['publishedAt'] = pd.to_datetime(covid['publishedAt'])
df['publishedAt'] = pd.to_datetime(df['publishedAt'])
```

```python
# merge 
video = df.merge(covid, how = 'left', on = 'publishedAt')
```

```python
# export dataset 
video.to_csv('video', index = False)
```
