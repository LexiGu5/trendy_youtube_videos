```python
import pandas as pd
```

```python
# load the data files 
df = pd.read_csv('trending_youtube.csv')
covid = pd.read_csv('covid.csv')
```

```python
# rename the column for merging 
covid.rename(columns = {'Date':'trending_date'}, inplace = True)
```

```python
# convert object to datetime to make the data type consistent 
covid['trending_date'] = pd.to_datetime(covid['trending_date'])
df['trending_date'] = pd.to_datetime(df['trending_date'])
```

```python
# merge 
video = df.merge(covid, how = 'left', on = 'trending_date')
```

```python
# look at the df
video
```

```python
# export dataset 
video.to_csv('video.csv', index = False)
```
